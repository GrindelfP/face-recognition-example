```python
from fer import FER
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
neutral_image = plt.imread("photos/neutral.jpg")
happy_image = plt.imread("photos/happy.jpg")
joy_image = plt.imread("photos/joyful.jpg")
sleepy_image = plt.imread("photos/sleepy.jpg")
irritated_image = plt.imread("photos/irritated.jpg")
fear_image = plt.imread("photos/fear.jpeg")
very_happy_image = plt.imread("photos/very-happy.jpeg")
disgust_image = plt.imread("photos/disgust.jpeg")
sadness_image = plt.imread("photos/sadness.jpeg")
surprized_image = plt.imread("photos/surprized.jpeg")
```

```python
emotions_detector = FER(mtcnn=True)
```

# 1. Нейтральные эмоции

```python
print("Intended emotion - neutral")
captured_emotions = emotions_detector.detect_emotions(neutral_image)
print(captured_emotions)
plt.imshow(neutral_image)
dominant_emotion, emotion_score = emotions_detector.top_emotion(neutral_image)
print(f"The dominant emotion is {dominant_emotion}: {emotion_score}")
```

Модель смогла довольно точно распознать нейтральные эмоции.

# 2. Счастье

```python
print("Intended emotion - happy")
captured_emotions = emotions_detector.detect_emotions(happy_image)
print(captured_emotions)
plt.imshow(happy_image)
dominant_emotion, emotion_score = emotions_detector.top_emotion(happy_image)
print(f"The dominant emotion is {dominant_emotion}: {emotion_score}")
```

Модель смогла довольно точно распознать счастье, причем точнее, чем нейтральные эмоции.

# 3. Радость

```python
print("Intended emotion - joyful")
captured_emotions = emotions_detector.detect_emotions(joy_image)
print(captured_emotions)
plt.imshow(joy_image)
dominant_emotion, emotion_score = emotions_detector.top_emotion(joy_image)
print(f"The dominant emotion is {dominant_emotion}: {emotion_score}")
```

Модель ошиблась в распознании радости, приняв ее за нейтральные эмоции, причем даже с большей вероятностью чем, собственно, нейтральные эмоции.
